#### Figure S17 A-D in M&S (2020) AS&T

This figure is located in the supplemental information (SI) of the paper.

The figure illustrates the decomposition of psd of the test dust samples into log-normal distributions to determine how many distribution modes exist in a test dust. Once the log-normal distribution modes were set (by trial and error estimations), the summary of all distributions would be calculated and compared to the measured psd of the test dust to verify how the estimations were accurate (which in this case, it shows they were).

The illustrations in this figure attempt to support the hypothesis that any dust or other particulate matter (PM) sample can be broken into smaller samples all of which have log-normal beahaviours.

Plot types: Line plot

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

# exec(open('notion_corrections.py').read())
df = pd.read_excel(r'dust_psd_50_iso_00a.xlsx', skiprows=9)
df = df[['Size', 'Mode 1', 'Mode 2', 'Mode 3']]
df.head()
```

```python
####################
### Figure S17 A ###
####################

plt.figure()
plt.plot(df['Size'], df['Mode 1'], c = "k", label = 'Mode 1: GM=0.9 µm, GSD=1.2')
plt.plot(df['Size'], df['Mode 2'], c = "dimgray", label = 'Mode 2: GM=4.3 µm, GSD=2.5')
plt.plot(df['Size'], df['Mode 3'], c = "gainsboro", label = 'Mode 3: GM=32 µm, GSD=2')

plt.xscale('log')

a = [0.3, 1, 3, 10, 30, 100, 300]

plt.xlim(0.3, 700)
plt.xticks(ticks = a, labels = a)
plt.minorticks_off()
plt.xlabel('Particle Size (µm)', size = 14)

plt.yticks(np.arange(0, 6, step=1))
plt.ylim(0, 5)
plt.ylabel('Volume Distribution (%)', size = 14)

plt.tick_params(axis = 'both', direction = 'inout')

plt.legend(loc = 2, frameon = True, edgecolor = 'black', framealpha=1, prop={'size': 8})
plt.savefig(r'FigureS17_iso_single.jpg', format='jpg', dpi=1600, bbox_inches='tight')
```

```python
### Data preparation for Figure S17B

df = pd.read_excel(r'dust_psd_50_iso_00a.xlsx', skiprows=9)

df['psd'] = df['Mode 1'] + df['Mode 2'] + df['Mode 3']
df = df[['Size', 'psd', 'Real']]
df.iloc[10:,:].head(10)
```

```python
####################
### Figure S17 B ###
####################

plt.figure()
plt.plot(df['Size'], df['Real'], c = "b", linewidth = 1, label = 'Measured (LDPS)')
plt.plot(df['Size'], df['psd'], c = "r", linestyle = '--', linewidth = 1, label = 'Fitted')

plt.xscale('log')

a = [0.3, 1, 3, 10, 30, 100, 300]

plt.xlim(0.3, 700)
plt.xticks(ticks = a, labels = a)
plt.minorticks_off()
plt.xlabel('Particle Size (µm)', size = 14)

plt.yticks(np.arange(0, 6, step=1))
plt.ylim(0, 5)
plt.ylabel('Volume Distribution (%)', size = 14)

plt.tick_params(axis = 'both', direction = 'inout')
plt.legend(loc = 1, frameon = True, edgecolor = 'black', framealpha=1, prop={'size': 10})
plt.savefig(r'FigureS17_iso_combined.jpg', format='jpg', dpi=1600, bbox_inches='tight')
```
