#### Figures S19 (A-K) and S20 (A-I) in M & S (2020) AS&T

This figure illustrates the outcome of modal analysis PSDs fitted best using the measured PSDs in all the experiments.
Due to their similarity, all plots are created using a for loop in an iterative process

Plot type: Line plot

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import os
import glob

# exec(open('notion_corrections.py').read())
os.chdir((r'C:\Users\alima\PhD Codes\Paper 1 Extraction\Raw\artificial_lds\modal'))
    
file_list = glob.glob('dust*.xlsx')

for file in file_list:
    if file[-11:-10] != '5':
        df = pd.read_excel(file, skiprows = 9)
        df = df[['Size', 'Made', 'Real']]
        
        plt.figure()
        plt.plot(df['Size'], df['Real'], c = "k", linewidth = 1, label = 'Measured - LDPS')
        plt.plot(df['Size'], df['Made'], c = "k", linestyle = '--', linewidth = 1, label = 'Fitted')

        plt.xscale('log')

        plt.yticks(np.arange(0, 6, step=1))
        plt.ylim(0, 5)
        plt.ylabel('Volume Distribution (%)', size = 14)

        plt.tick_params(axis = 'both', direction = 'inout')
        plt.legend(loc = 2, frameon = True, edgecolor = 'black', framealpha=1, prop={'size': 10})
        
        if file[-12:-9] == 'iso':
            a = [0.3, 1, 3, 10, 30, 100, 300]

            plt.xlim(0.3, 700)
            plt.xticks(ticks = a, labels = a)
            plt.minorticks_off()
            plt.xlabel('Particle Size (µm)', size = 14)
            
        elif file[-12:-9] == 'ash':
            a = [0.3, 1, 3, 10, 30, 100, 300, 1000, 3000]

            plt.xlim(0.3, 7300)
            plt.xticks(ticks = a, labels = a)
            plt.minorticks_off()
            plt.xlabel('Particle Size (µm)', size = 14)
            
            
        plt.savefig(r'FigureS20_' + file[-15:-13] + '.jpg', format='jpg', dpi=1600, bbox_inches='tight')
```
