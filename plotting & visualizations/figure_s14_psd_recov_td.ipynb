#### Figure S14 A-C in M&S (2020) AS&T

The plot is in the supplemental information (SI) of the paper.

This plot indicates how PSDs of the recovered dust from different filters loaded with different masses of dust differ from the PSD of the test dust by which the filters were loaded.

Plot type: Line plot

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_excel(r'artl_v_psd_master.xlsx')
col_list_all = list(df.columns)

# Picking standardized test dust sample PSDs 
df_iso = df['Eval_010_iso2_11e_td_180326_am_mean']
df_ash = df['Eval_001_ash2_11e_td_171128_am_mean']

# Selecting those relevant samples
df_iso_08e = df[[col for col in col_list_all if ('_iso' in col) & ('_08e' in col) & ('_hd' in col) & ('_mean' in col)]]
df_iso_11e = df[[col for col in col_list_all if ('_iso2' in col) & ('_11e' in col) & ('_hd' in col) & ('_mean' in col) & ('r2' not in col)]]
df_ash_11e = df[[col for col in col_list_all if ('_ash' in col) & ('_11e' in col) & ('_hd' in col) & ('_mean' in col)]]

for item in [df_iso_08e, df_iso_11e, df_ash_11e]:
    for col in item.columns:
        item.rename(columns = {col:col[4:8]}, inplace = True)
        
df_iso_11e = df_iso_11e[['_010', '_014', '_003', '_013', '_012', '_007', '_006']]
df_ash_11e = df_ash_11e[['_001', '_004', '_002', '_005', '_015', '_009']]

# labelling and color codes
color_code = {1: 'green',
              2: 'mediumblue',
              3: 'blue',
              4: 'royalblue',
              5: 'cornflowerblue',
              6: 'lightsteelblue',
              7: 'lavender'}

iso_8e_label = {3: 'Recovered (1.2 g)',
                7: 'Recovered (0.8 g)'}

iso_label = {1: 'Recovered (13.6 g)',
             2: 'Recovered (3.4 g)',
             3: 'Recovered (3.3 g)',
             4: 'Recovered (2.2 g)',
             5: 'Recovered (1.3 g)',
             6: 'Recovered (1.0 g)',
             7: 'Recovered (0.6 g)'}

ash_label = {1: 'Recovered (14.8 g)',
             2: 'Recovered (3.6 g)',
             3: 'Recovered (3.6 g)',
             4: 'Recovered (3.6 g)',
             5: 'Recovered (2.0 g)',
             6: 'Recovered (0.8 g)'}
```

```python
#####################
#### Figure 14 a ####
#####################

plt.figure()
plt.plot(df['Size'], df_iso, c = 'red', label = 'Test Dust: ISO Fine')

j = 0
for i in iso_8e_label:
    plt.plot(df['Size'], df_iso_08e.iloc[:,j], c = color_code[i], label = iso_8e_label[i], linestyle = '--')
    j += 1

plt.xscale('log')
     
plt.xlim(0.3, 3000)
a = [0.3, 1, 3, 10, 30, 100, 300, 1000, 3000]
plt.xticks(ticks = a, labels = a)
plt.minorticks_off()
plt.xlabel('Particle Size (µm)', size = 14)

plt.ylim(0, 5)
plt.yticks(np.arange(0, 6, step=1))
plt.ylabel('Volume Distribution (%)', size = 14)

plt.tick_params(axis = 'both', direction = 'inout')

plt.legend(loc = 1, frameon = True, edgecolor = 'black', framealpha=1, fontsize=9)
plt.savefig(r'FigureS14a.jpg', format='jpg', dpi=1600, bbox_inches='tight')
```

```python
#####################
#### Figure 14 b ####
#####################

plt.figure()
plt.plot(df['Size'], df_iso, c = 'red', label = 'Test Dust: ISO Fine')

for i in iso_label:
    plt.plot(df['Size'], df_iso_11e.iloc[:,i-1], c = color_code[i], label = iso_label[i], linestyle = '--')

plt.xscale('log')
     
plt.xlim(0.3, 3000)
a = [0.3, 1, 3, 10, 30, 100, 300, 1000, 3000]
plt.xticks(ticks = a, labels = a)
plt.minorticks_off()
plt.xlabel('Particle Size (µm)', size = 14)

plt.ylim(0, 5)
plt.yticks(np.arange(0, 6, step=1))
plt.ylabel('Volume Distribution (%)', size = 14)

plt.tick_params(axis = 'both', direction = 'inout')

plt.legend(loc = 1, frameon = True, edgecolor = 'black', framealpha=1, fontsize=9)
plt.savefig(r'FigureS14b.jpg', format='jpg', dpi=1600, bbox_inches='tight')
```

```python
#####################
#### Figure 14 c ####
#####################

plt.figure()
plt.plot(df['Size'], df_ash, c = 'red', label = 'Test Dust: ASHRAE #2')

for i in ash_label:
    plt.plot(df['Size'], df_ash_11e.iloc[:,i-1], c = color_code[i], label = ash_label[i], linestyle = '--')

plt.xscale('log')
     
plt.xlim(0.3, 3000)
a = [0.3, 1, 3, 10, 30, 100, 300, 1000, 3000]
plt.xticks(ticks = a, labels = a)
plt.minorticks_off()
plt.xlabel('Particle Size (µm)', size = 14)

plt.ylim(0, 5)
plt.yticks(np.arange(0, 6, step=1))
plt.ylabel('Volume Distribution (%)', size = 14)

plt.tick_params(axis = 'both', direction = 'inout')

plt.legend(loc = 1, frameon = True, edgecolor = 'black', framealpha=1, fontsize=9)
plt.savefig(r'FigureS14c.jpg', format='jpg', dpi=1600, bbox_inches='tight')

```
