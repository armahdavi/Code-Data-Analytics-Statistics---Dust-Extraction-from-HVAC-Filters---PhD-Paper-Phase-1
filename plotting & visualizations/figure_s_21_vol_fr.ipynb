#### Figure S21 A,B in M & S (2020) AS & T

This figure illustrated the volume fractions of test dust along with those averaged from the PSDs of the recovered dust. More explanations are in the figure.

Plot type: Normalized stacked bar

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

## Hard coding for the barriers of vol. fractions of ISO and ASHRAE test dust

iso_1 = 0.021645
iso_2 = 0.675325 + 0.021645

ash_1 = 0.0163577
ash_2 = 0.51036 + 0.0163577
ash_3 = 0.229008 + 0.51036 + 0.0163577
ash_4 = 0.0697928 + 0.229008 + 0.51036 + 0.0163577
ash_5 = 0.0610687 + 0.0697928 + 0.229008 + 0.51036 + 0.0163577
```

```python
## Preparation code for Figure S21 A (ISO sample vol fractions)

# exec(open('notion_corrections.py').read())
df = pd.read_excel(r'psd_modal_analysis_summary.xlsx')
# df = df[(df['td'] == 1) & (df['ft'] != 0)] # why not working
# dustmass = pd.read_excel(r'artl_dataset_summary_collapsed.xlsx')

# erging with loaded dust mass in filter (noted ast "dustmass" in the data frame) for comparison purposes
# df.rename(columns = {'expn':'ExpN'}, inplace = True)
# df = df.merge(dustmass[['ExpN', 'dustmass']], on = 'ExpN', how= 'left')

df = df[df['ft'] != 0] 
df = df[df['td'] == 1]

df.sort_values(['ft', 'dustmass'], inplace = True)

x = ['MERV 8', 'MERV 8E', 'MERV 8E', 'MERV 11', 'MERV 11', 'MERV 11', 'MERV 11', 'MERV 11', 'MERV 11', 'MERV 11', 'MERV 14']
v = list(np.arange(1,12,1))

label_iso = dict(zip(v,x))

x2 = list(np.arange(0.9,14,1.2))
label_iso = dict(zip(x2,x))
```

```python
df.head()
```

```python
####################
### Figure S21 A ###
####################

plt.figure()

plt.bar(x2, df['mode_1_f'], color='k', label = 'Mode 1: GM=0.9 µm, GSD=1.2', edgecolor = "k", linewidth = 0.8)
plt.bar(x2, df['mode_2_f'], bottom=df['mode_1_f'], color='grey', label = 'Mode 2: GM=4.3 µm, GSD=2.5', edgecolor = "k", linewidth = 0.8)
plt.bar(x2, df['mode_3_f'], bottom=df['mode_1_f'] + df['mode_2_f'], color='lightgrey', label = 'Mode 3: GM=32 µm, GSD=2', edgecolor = "k", linewidth = 0.8)

plt.xticks([])
plt.yticks(np.arange(0, 1.1, step = 0.2))
plt.ylabel('Volume Fraction (-)', size = 14)
plt.tick_params(axis = 'y', direction = 'inout')

for i in v:
    plt.text(list(x2)[i-1]+0.05, -0.05, list(label_iso.values())[i-1], fontsize = 6, ha='center')

plt.axhline(y = iso_1, linewidth=0.8, color = 'k', linestyle = (0, (8, 9)))
plt.axhline(y = iso_2, linewidth=0.8, color = 'k', linestyle = (0, (8, 9)), alpha = 0.5)

plt.xlim(0.2, 13.6)
plt.ylim(0, 1)

plt.legend(loc = 2, frameon = True, edgecolor = 'black', framealpha=1, bbox_to_anchor =(0.29, -0.08), prop={'size': 7})

plt.savefig(r'FigureS21A.jpg', format='jpg', dpi=1600, bbox_inches='tight')
```

```python
## Preparation code for Figure S21 B (ASHRAE sample vol fractions)

df = pd.read_excel(r'psd_modal_analysis_summary.xlsx')
# df = df[(df['td'] == 1) & (df['ft'] != 0)] # why not working
# dustmass = pd.read_excel(r'artl_dataset_summary_collapsed.xlsx')
# df.rename(columns = {'expn':'ExpN'}, inplace = True)
# df = df.merge(dustmass[['ExpN', 'dustmass']], on = 'ExpN', how= 'left')
df = df[df['ft'] != 0] 
df = df[df['td'] == 2]

df.sort_values(['ft', 'dustmass'], inplace = True)

x = ['MERV 8', 'MERV 8E', 'MERV 11', 'MERV 11', 'MERV 11', 'MERV 11', 'MERV 11', 'MERV 11', 'MERV 14']
v = list(np.arange(1,10,1))

x2 = list(np.arange(0.8,11.5,1.2))
label_ash = dict(zip(x2,x))
```

```python
####################
### Figure S21 B ###
####################

plt.figure()

plt.bar(x2, df['mode_1_f'], color='k', label = 'Mode 1: GM=0.9 µm, GSD=1.2', edgecolor = "k", linewidth = 0.8)
plt.bar(x2, df['mode_2_f'], bottom=df['mode_1_f'], color='grey', label = 'Mode 2: GM=4.3 µm, GSD=2.5', edgecolor = "k", linewidth = 0.8)
plt.bar(x2, df['mode_3_f'], bottom=df['mode_1_f'] + df['mode_2_f'], color='lightgrey', label = 'Mode 3: GM=32 µm, GSD=2', edgecolor = "k", linewidth = 0.8)

plt.bar(x2, df['mode_4_f'], bottom=df['mode_1_f'] + df['mode_2_f'] + df['mode_3_f'], 
        facecolor='#653700', alpha = 0.8, label = 'Mode 4: GM=20 µm, GSD=1.5', edgecolor = "k", linewidth = 0.8)
plt.bar(x2, df['mode_5_f'], bottom=df['mode_1_f'] + df['mode_2_f'] + df['mode_3_f'] + df['mode_4_f'], 
        color='#D2B48C', label = 'Mode 5: GM=150 µm, GSD=2', edgecolor = "k", linewidth = 0.8)
plt.bar(x2, df['mode_6_f'], bottom=df['mode_1_f'] + df['mode_2_f'] + df['mode_3_f'] + df['mode_4_f'] + df['mode_5_f'], 
        color='linen', label = 'Mode 6: GM=550 µm, GSD=1.6', edgecolor = "k", linewidth = 0.8)

plt.xticks([])
plt.yticks(np.arange(0, 1.1, step = 0.2))
plt.ylabel('Volume Fraction (-)', size = 14)
plt.tick_params(axis = 'y', direction = 'inout')

for i in v:
    plt.text(list(x2)[i-1]+0.05, -0.05, list(label_iso.values())[i-1], fontsize = 6, ha='center')

plt.axhline(y = ash_1, linewidth=0.8, color = 'k', linestyle = (0, (8, 9)))
plt.axhline(y = ash_2, linewidth=0.8, color = 'k', linestyle = (0, (8, 9)), alpha = 0.5)
plt.axhline(y = ash_3, linewidth=0.8, color = 'k', linestyle = (0, (8, 9)), alpha = 0.2)
plt.axhline(y = ash_4, linewidth=0.8, color = '#653700', linestyle = (0, (8, 9)))
plt.axhline(y = ash_5, linewidth=0.8, color = '#653700', linestyle = (0, (8, 9)), alpha = 0.5)

plt.xlim(0.2, 11)
plt.ylim(0, 1)

plt.legend(loc = 2, frameon = True, edgecolor = 'black', framealpha=1, bbox_to_anchor =(0.071, -0.08), ncol = 2, prop={'size': 7})

plt.savefig(r'FigureS21B.jpg', format='jpg', dpi=1600, bbox_inches='tight')
```
